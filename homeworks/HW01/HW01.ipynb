# Примеры создания массивов

```python
lst = [1,3,2,4,1,5]
array_from_list = np.array(lst)

a = np.arange(10)
b = np.ones(2)
c = np.zeros(3)
# соответственно функции создания массива из списка, с последовательными числами, заполненный нулями и единицами.
array_from_list,a,b,c
```

# Демонстрация атрибутов

```python
a = np.array([[1,2,3],
              [3,5,6],
              [1,7,9]])
# вывод формы массива и типа значений его элементов
a.shape, a.dtype 
```

# Примеры индексирования и срезов

```python
b = a[1] # второй подмассив
c = a[-1][-1] # последний элемент двумерного массива - 9 (число)
g = a[0][:2] # из первого подмассива элементы с 0 по 2 невключительн

b,c,g
```

# Векторные операции и базовые статистики

```python
a = np.array([2,3,5])
b = np.array([1,7,9])

c = a * b # поэлементное умножение
g = a + b #поэлементное сложение
f = a @ b.T  # матричное умножение
j = np.stack([a,b],axis = 0) # конкатенация

c, g, f,j
```

```python
m = j.mean()
v = j.var(axis = 0)
s = j.std(axis = 0)
# расчет среднего, дисперсии и стандартного отклонения
m,v,s
```
